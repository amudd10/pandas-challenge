```python
total_student_per_school = combined_school_df["school_name"].value_counts()
total_student_per_school

#Total budget per school
budget_per_school = combined_school_df.groupby(["school_name"]).mean()["budget"]
budget_per_school

per_student_budget = budget_per_school/total_student_per_school
per_student_budget

#avg math score
avg_math_score_per_school = round(combined_school_df.groupby(["school_name"]).mean()["math_score"], 1)
avg_math_score_per_school

#avg read score
avg_read_score_per_school = round(combined_school_df.groupby(["school_name"]).mean()["reading_score"],1)
avg_read_score_per_school

# % math passing per school

#make sure these calculations are right
students_pass_math = combined_school_df[combined_school_df["math_score"] >=70]
students_pass_math_per_school = students_pass_math.groupby(["school_name"]).count()["student_name"]
students_pass_math_per_school

#perc_of_students_pass_math_per_school = (students_pass_math_per_school/total_student_per_school) *100
#perc_of_students_pass_math_per_school
perc_of_students_pass_math_per_school = round((students_pass_math_per_school/total_student_per_school) *100,1)
perc_of_students_pass_math_per_school

# % reading passing per school
students_pass_reading = combined_school_df[combined_school_df["reading_score"] >=70]
students_pass_reading_per_school = students_pass_reading.groupby(["school_name"]).count()["student_name"]
students_pass_reading_per_school

perc_of_students_pass_reading_per_school = round((students_pass_reading_per_school/total_student_per_school) *100,1)
perc_of_students_pass_reading_per_school

#% of Overall passing students(passed both math and reading)

#do .loc research for these 2 conditionals
overall_passing_perc_per_school = (perc_of_students_pass_reading_per_school + perc_of_students_pass_math_per_school)/2
overall_passing_perc_per_school

#School Summary Dataframe
# school_summary_df = pd.DataFrame(index = school_name)

# school_type = {}
# total_students = {}
# total_student_budget = {}
# per_student_budget = {}
# avg_math_score_per_school = {}
# avg_read_score_per_school = {}

# school_summary_df["School Type"] = school_data["type"].tolist()
per_school_summary_df = pd.DataFrame({
                                  "School type":school_type,
                                  "Total students":total_student_per_school,
                                  "Total school budget":budget_per_school,
                                  "Per student budget":per_student_budget,
                                  "Average math score":avg_math_score_per_school,
                                  "Average reading score":avg_read_score_per_school,
                                  "% passing math":perc_of_students_pass_math_per_school,
                                  "% passing reading":perc_of_students_pass_reading_per_school,
                                  "% overall passing":overall_passing_perc_per_school
                                 }
                                )
per_school_summary_df
```

```python
#Math Scores by Grade
ninth_grade_math = student_data.loc[student_data['grade'] == '9th'].groupby('school_name')["math_score"].mean()
tenth_grade_math =  student_data.loc[student_data['grade'] == '10th'].groupby('school_name')["math_score"].mean()
eleventh_grade_math =  student_data.loc[student_data['grade'] == '11th'].groupby('school_name')["math_score"].mean()                                                                             
twelfth_grade_math =   student_data.loc[student_data['grade'] == '12th'].groupby('school_name')["math_score"].mean()

#Creating Dataframe
math_scores_by_grade = pd.DataFrame({
                                        "9th": ninth_grade_math,
                                        "10th": tenth_grade_math,
                                        "11th": eleventh_grade_math,
                                        "12th": twelfth_grade_math
                                    })
                                                                                 
math_scores_by_grade                                                                                
```

```python
#Reading Scores by Grade
ninth_grade_reading = student_data.loc[student_data['grade'] == '9th'].groupby('school_name')["reading_score"].mean()
tenth_grade_reading =  student_data.loc[student_data['grade'] == '10th'].groupby('school_name')["reading_score"].mean()
eleventh_grade_reading =  student_data.loc[student_data['grade'] == '11th'].groupby('school_name')["reading_score"].mean()                                                                             
twelfth_grade_reading =   student_data.loc[student_data['grade'] == '12th'].groupby('school_name')["reading_score"].mean()

#Creating Dataframe
reading_scores_by_grade = pd.DataFrame({
                                        "9th": ninth_grade_reading,
                                        "10th": tenth_grade_reading,
                                        "11th": eleventh_grade_reading,
                                        "12th": twelfth_grade_reading
                                    })
                                                                                 
reading_scores_by_grade        
```

```python
#Scores by School Spending
#Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following

bins = [0,585, 630, 645, 680]
spending_categories = ["<$585", "$585-630", "$630-645", "$645-680"]

combined_school_df["Spending Ranges (Per Student)"] = pd.cut(
    combined_school_df['budget']/combined_school_df['size'], bins, labels = spending_categories)

grouped_by_spending = combined_school_df.groupby('Spending Ranges (Per Student)')

#calculating avg's

avg_math = round(grouped_by_spending['math_score'].mean(),2)
avg_read = round(grouped_by_spending['reading_score'].mean(),2)
math_pass_percent = round(combined_school_df[combined_school_df['math_score'] >= 70].groupby('Spending Ranges (Per Student)')['student_name'].count()/grouped_by_spending['student_name'].count(),2)*100
reading_pass_percent = round(combined_school_df[combined_school_df['reading_score'] >= 70].groupby('Spending Ranges (Per Student)')['student_name'].count()/grouped_by_spending['student_name'].count(),2)*100
#and
overall_pass_percent = round(combined_school_df[(combined_school_df['reading_score'] >= 70) & (combined_school_df['math_score'] >= 70)].groupby('Spending Ranges (Per Student)')
                                            ['student_name'].count()/grouped_by_spending['student_name'].count(),2)*100

#Dataframe to hold bins
scores_by_school_spending = pd.DataFrame({
                                        "Average Math Score": avg_math,
                                        "Average Reading Score": avg_read,
                                        "% Passing Math": math_pass_percent,
                                        "% Passing Reading": reading_pass_percent,
                                        "% Overall Passing": overall_pass_percent      
                                        })

scores_by_school_spending
```

```python
# Scores by School Size
# Perform the same operations as above, based on school size.
bins = [0, 1000, 2000, 5000]
sizes = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
combined_school_df["School Size"] = pd.cut(combined_school_df["size"], bins, labels = sizes)

grouped_by_size = combined_school_df.groupby("School Size")

#calculating avg's

avg_math = round(grouped_by_size['math_score'].mean(),2)
avg_read = round(grouped_by_size['reading_score'].mean(),2)
math_pass_percent = round(combined_school_df[combined_school_df['math_score'] >= 70].groupby('School Size')['student_name'].count()/grouped_by_size['student_name'].count(),2)*100
reading_pass_percent = round(combined_school_df[combined_school_df['reading_score'] >= 70].groupby("School Size")['student_name'].count()/grouped_by_size['student_name'].count(),2)*100
#and
overall_pass_percent = round(combined_school_df[(combined_school_df['reading_score'] >= 70) & (combined_school_df['math_score'] >= 70)].groupby("School Size")
                                            ['student_name'].count()/grouped_by_size['student_name'].count(),2)*100
                                                                                                                                                 

#Dataframe to hold bins
scores_by_school_size = pd.DataFrame({
                                        "Average Math Score": avg_math,
                                        "Average Reading Score": avg_read,
                                        "% Passing Math": math_pass_percent,
                                        "% Passing Reading": reading_pass_percent,
                                        "% Overall Passing": overall_pass_percent      
                                        })

scores_by_school_size
```

```python
# Scores by School Type
# Perform the same operations as above, based on school type

score_by_type = combined_school_df.groupby("type")

#calculating avg's

avg_math = round(score_by_type['math_score'].mean(),2)
avg_read = round(score_by_type['reading_score'].mean(),2)
math_pass_percent = round(combined_school_df[combined_school_df['math_score'] >= 70].groupby('type')['student_name'].count()/score_by_type['student_name'].count(),2)*100
reading_pass_percent = round(combined_school_df[combined_school_df['reading_score'] >= 70].groupby("type")['student_name'].count()/score_by_type['student_name'].count(),2)*100
#and
overall_pass_percent = round(combined_school_df[(combined_school_df['reading_score'] >= 70) & (combined_school_df['math_score'] >= 70)].groupby("type")
                                            ['student_name'].count()/score_by_type['student_name'].count(),2)*100

#Dataframe to hold bins
scores_by_school_type.index.name = "Type of School"
scores_by_school_type = pd.DataFrame({
                                        "Average Math Score": avg_math,
                                        "Average Reading Score": avg_read,
                                        "% Passing Math": math_pass_percent,
                                        "% Passing Reading": reading_pass_percent,
                                        "% Overall Passing": overall_pass_percent      
                                        })

scores_by_school_type
```
